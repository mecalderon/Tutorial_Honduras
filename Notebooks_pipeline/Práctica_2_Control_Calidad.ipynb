# **Capacitación de secueciación genómica y análisis bioinformático**

## **Instructivo de comandos**

## **Practica 2. Control de Calidad**

Este instructivo tiene como objetivo que cada participante pueda evaluar la calidad de las secuencias generadas durante el taller.

## **¿Qué hacer con los archivos Fastq tras secuenciar una muestra de Salmonella?**

1. **Control de calidad (QC)**: **FastQC**, TrimGalore, Kraken2.

2. **Ensamblaje de genomas**: Unicycler.

3. **Caracterización y tipificación**: identificación de determinantes de resistencia mediante SeqSero2 y AMRFinderPlus.

4. **Análisis filogenético**: construcción de relaciones evolutivas entre las muestras.

Este instructivo se centra específicamente en el uso de **FastQC** para el control de calidad de las secuencias.

Este cuaderno está diseñado para ejecutar **FastQC** directamente desde la **terminal**.

## 1. **Ejecución de FastQC desde la terminal**

Ingrese a la terminal seleccionando el ícono correspondiente:

![Ubuntu](https://raw.githubusercontent.com/mecalderon/Tutorial_Honduras/main/Figures/Ubuntu.png)


Al ingresar a la terminal cree la carpeta **Fastq** donde se almacenarán los archivos de secuencia:
```
mkdir Fastq
```

Se generarán accesos directos a los archivos FASTQ que se analizarán dentro de la carpeta **Fastq**. En esta práctica, cada participante evaluará la calidad de dos muestras. Para cada una, es necesario cargar ambos archivos de lectura: R1 y R2.

Para crear los accesos directos, utilice el siguiente comando. Reemplace **$DIR** por el directorio donde se encuentran las secuencias y **XXXX** por el número correspondiente de la muestra:
```
ln -s $DIR/XXXX_S*_L001_R*_001.fastq.gz Fastq
```


Renombre los archivos Fastq:
```
for f in Fastq/*_S*_L001_R1_001.fastq.gz; do mv $f ${f%_S*_L001_R1_001.fastq.gz}_1.fastq.gz; done

for f in Fastq/*_S*_L001_R2_001.fastq.gz; do mv $f ${f%_S*_L001_R2_001.fastq.gz}_2.fastq.gz; done
```

Cree una carpeta para almacenar los resultados de FastQC:
```
mkdir 2-Practica_Fastqc
```

Active el entorno de conda necesario para ejecutar la herramienta:
```
source activate microenv
```

Ejecute FastQC sobre los archivos R1:
```
fastqc Fastq/*_1.fastq.gz -o 2-Practica_Fastqc
```

Ejecute FastQC sobre los archivos R2:
```
fastqc Fastq/*_2.fastq.gz -o 2-Practica_Fastqc
```

Descargue desde la carpeta **2-Practica_Fastqc** los reportes de calidad de las muestras. Los archivos a analizar son los **.html** generados por FastQC.
