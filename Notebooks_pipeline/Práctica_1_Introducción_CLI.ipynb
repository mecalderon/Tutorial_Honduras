## **Practica 1. Introducción a Linux y línea de comandos**

El presente instructivo tiene como objetivo familiarizar al usuario con el uso de la terminal y la ejecución de comandos básicos en Linux.

### **1. Uso de WSL**


Para utilizar WSL, escriba Ubuntu en el buscador de Windows y abra la aplicación.

![Ubuntu](https://raw.githubusercontent.com/mecalderon/Tutorial_Honduras/main/Figures/Ubuntu.png)

Este cuaderno permite ejecutar comandos básicos de Linux.

Inicialmente, todos los participantes deberán clonar el repositorio de **GitHub** que contiene los notebooks que se utilizarán durante la capacitación.

Para ello, abra la terminal e ingrese al directorio **Desktop**:
```
cd Desktop
```

A continuación, clone en su directorio el repositorio que contiene los notebooks:
```
git clone https://github.com/mecalderon/Tutorial_Honduras.git
```

## **2. Uso de la terminal**

En esta sección trabajaremos directamente desde la **terminal** para ejecutar los comandos.

Obtenga la ruta del directorio actual ejecutando el siguiente comando:
```
pwd
```
Visualice el contenido del directorio:
```
ls
```  
Cree una carpeta llamada **1-Practica_Linux**:

```
mkdir 1-Practica_Linux
```

Verifique que el directorio fue creado correctamente:
```
ls -lrt
```

Ingrese al directorio **1-Practica_Linux**:
```
cd 1-Practica_Linux
```

Obtenga la ruta del directorio actual:
```
pwd
```

Cree la carpeta **nacionalidad**:

```
mkdir nacionalidad  
```

Ingrese al directorio recién creado:
```
cd nacionalidad
```

Cree un archivo de texto llamado **nombre** y verifique que fue creado:
```
touch nombre

ls -lrt
```

Utilice un editor de texto para modificar el archivo creado:

```
nano nombre  
```
**Escriba su nombre dentro del archivo**

Para salir del editor:
```
Ctrl + X
Y (yes / sí para guardar)
Enter
```

Visualice el contenido del archivo que acaba de crear:
```
cat nombre
```

Cree un archivo de texto llamado **pais**. (Opcionalmente, puede usar las flechas del teclado para recuperar el comando anterior y modificarlo). Luego verifique que el archivo fue creado:
```
touch pais

ls  -lrt
```

Utilice un editor de texto para modificar el archivo:
```
nano pais
```
**Escriba su nacionalidad dentro del archivo**

Para salir del editor:
```
Ctrl + X
Y (yes / sí para guardar)
Enter
```

Visualice el contenido del archivo creado:
```
cat pais
```

Concatene ambos archivos en uno nuevo llamado **informacion** y verifique que fue creado:
```
cat nombre pais > informacion

ls -lrt
```

Visualice el contenido del nuevo archivo:
```
cat informacion
```

Regrese al directorio **1-Practica_Linux**:  
```
cd ..
```

Realice un acceso directo al archivo **informacion** y explore su contenido:
```
ln -s $PWD/nacionalidad/informacion .  

ls -lrt  

cat informacion
```

Elimine la carpeta **nacionalidad** y el acceso directo del archivo **informacion**, y verifique que ambos fueron eliminados:
```
rm -r nacionalidad

ls -lrt

rm *

ls -lrt  
```

Cree el directorio **fastq** e ingrese en él:
```
mkdir fastq  

cd fastq
```

Cree accesos directos a los archivos de secuencias:
```
 ln -s ../../Tutorial_Honduras/Fastq/*.fastq.gz .
```

Renombre los archivos fastq utilizando los siguientes comandos:
```
for f in *_S*_L001_R1_001.fastq.gz; do mv $f ${f%_S*_L001_R1_001.fastq.gz}_1.fastq.gz; done

for f in *_S*_L001_R2_001.fastq.gz; do mv $f ${f%_S*_L001_R2_001.fastq.gz}_2.fastq.gz; done
```

Compruebe que los nombres de los archivos cambiaron correctamente:
```
ls -lrth
```
Para salir de la terminal se ejecuta el siguiente comando:
```
exit
```
