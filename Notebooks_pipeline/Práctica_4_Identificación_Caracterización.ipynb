# **Capacitación de secueciación genómica y análisis bioinformático**

## **Instructivo de comandos**

## **Practica 4. Identificación y/o caracterización de cepas bacterianas (Salmonella)**

Este instructivo tiene como objetivo que cada participante realice la identificación y caracterización de cepas de *Salmonella* a partir de datos de secuenciación.

Para ello, se utilizará la herramienta **ARIBA** con el objetivo de identificar genes de resistencia a antibióticos y realizar la tipificación multilocus (MLST). Asimismo, la herramienta **SeqSero2** se empleará para determinar el serotipo de las cepas analizadas. Ambas herramientas utilizarán como entrada las lecturas de secuenciación previamente filtradas y recortadas (reads trimmeados).

Por otra parte, se utilizará la herramienta **AMRFinderPlus** para identificar genes de resistencia a antibióticos a partir de los genomas ensamblados.

## **¿Qué hacer con los archivos Fastq tras secuenciar una muestra de Salmonella?**

El análisis bioinformático de datos de secuenciación generalmente incluye los siguientes pasos:

1. **Control de calidad (QC)**: FastQC, TrimGalore, Kraken2.

2. **Ensamblaje de genomas**: Unicycler.

3. **Caracterización y tipificación**: identificación de determinantes de resistencia mediante **SeqSero2** y **AMRFinderPlus**.

4. **Análisis filogenético**: construcción de relaciones evolutivas entre las muestras.

Este instructivo se centra específicamente en el uso de **ARIBA**, **SeqSero2** y **AMRFinderPlus**.

### **Ejecución de ARIBA para AMR**

Active el entorno de conda necesario para ejecutar la herramienta:
```
source activate ariba_env
```

Cree la carpeta **1-ariba_amr**, donde se almacenarán los resultados generados por la herramienta:

```
mkdir 1-ariba_amr
```

**ARIBA** permite identificar genes de resistencia a antibióticos (AMR) directamente a partir de las lecturas de secuenciación. La anotación de estos genes puede variar dependiendo de la base de datos utilizada.

Por esta razón, en esta práctica se emplearán tres bases de datos diferentes: **ResFinder**, **CARD** y **VirulenceFinder**.

#### **Ejecución de ARIBA con la base de datos ResFinder**

Ejecute el siguiente comando para identificar genes de resistencia utilizando la base de datos **ResFinder**.

**Nota:** siempre se debe verificar la ruta donde se encuentra almacenada la base de datos correspondiente.

```
for f in ../3-Practica_Ensamblaje/1-trim_galore/*_1.fastq.gz;
do
base=$(basename ${f%_1.fastq.gz})

ariba run --threads 2 ../ariba_databases/resfinder.out/ \
$f ${f%_1.fastq.gz}_2.fastq.gz 1-ariba_amr/${base}_resfinder;

done
```

#### **Ejecución de ARIBA con la base de datos CARD**

Ejecute el siguiente comando utilizando la base de datos **CARD**.

**Nota:** siempre se debe verificar la ruta donde se encuentra almacenada la base de datos correspondiente.


```
for f in ../3-Practica_Ensamblaje/1-trim_galore/*_1.fastq.gz;
do
base=$(basename ${f%_1.fastq.gz})
    
ariba run --threads 2 ../ariba_databases/card.out/ \
$f ${f%_1.fastq.gz}_2.fastq.gz 1-ariba_amr/${base}_card;

done
```

#### **Ejecución de ARIBA con la base de datos VirulenceFinder**

Ejecute el siguiente comando utilizando la base de datos **VirulenceFinder**.

**Nota:** siempre se debe verificar la ruta donde se encuentra almacenada la base de datos correspondiente.

```
for f in ../3-Practica_Ensamblaje/1-trim_galore/*_1.fastq.gz;
do
base=$(basename ${f%_1.fastq.gz})
    
ariba run --threads 2 ../ariba_databases/vfd.out/ \
$f ${f%_1.fastq.gz}_2.fastq.gz 1-ariba_amr/${base}_vfdb;

done
```

La salida generada por **ARIBA** puede ser extensa, por lo que la herramienta permite procesar los resultados para generar un resumen más fácil de visualizar.

Ejecute el siguiente comando para resumir los resultados obtenidos con cada base de datos:
```
ariba summary 1-ariba_amr/resfinder_summary 1-ariba_amr/*_resfinder/report.tsv
ariba summary 1-ariba_amr/card_summary 1-ariba_amr/*_card/report.tsv
ariba summary 1-ariba_amr/vfdb_summary 1-ariba_amr/*_vfdb/report.tsv
```

Una vez completada la ejecución, encontrará en la carpeta **1-ariba_amr** los archivos **resfinder_summary.csv**, **card_summary.csv** y **vfdb_summary.csv**, que contienen los resultados consolidados para todas las muestras.

Abra uno de estos archivos y explórelo; tendrá una estructura similar a la mostrada en la imagen de referencia, lo que facilita la visualización y comparación de los genes de resistencia identificados.

![ARIBA_AMR](https://raw.githubusercontent.com/mecalderon/Tutorial_Panama/main/Figures/Ariba_AMR.png)


### **Ejecución de ARIBA para MLST**

Cree la carpeta **2-ariba_mlst**, donde se almacenarán los resultados generados por la herramienta:
```
mkdir 2-ariba_mlst
```
La herramienta **ARIBA** permite determinar el MLST directamente a partir de las lecturas de secuenciación.

Ejecute el siguiente comando para realizar el análisis de MLST con **ARIBA**.

**Nota:** siempre se debe verificar la ruta donde se encuentra almacenada la base de datos correspondiente.


```
for f in ../3-Practica_Ensamblaje/1-trim_galore/*_1.fastq.gz;
do
base=$(basename ${f%_1.fastq.gz})
    
ariba run --threads 2 ../ariba_databases/get_mlst/Salmonella_enterica_MLSTdb/ \
$f ${f%_1.fastq.gz}_2.fastq.gz 2-ariba_mlst/${base}_mlst;

done
```

Ejecute el siguiente comando para resumir los resultados obtenidos del MLST:
```
ariba summary 2-ariba_mlst/mlst_summary 2-ariba_mlst/*_mlst/report.tsv
```

Una vez completada la ejecución, en la carpeta **2-ariba_mlst** encontrará el archivo **mlst_summary.csv**, que contiene los resultados consolidados para todas las muestras analizadas.

Además, dentro de la carpeta **2-ariba_mlst** se generará una subcarpeta para cada muestra. En cada una de estas subcarpetas se encuentra el archivo **mlst_report.tsv**, que contiene los resultados detallados del análisis.

Abra este archivo y explórelo; observará una estructura similar a la mostrada en la imagen de referencia, donde se presentan los alelos identificados y el Sequence Type (ST) asignado a cada muestra.

![ARIBA_MLST](https://raw.githubusercontent.com/mecalderon/Tutorial_Panama/main/Figures/Ariba_MLST.png)


### **Ejecución de SeqSero2**

Para realizar la serotipificación de las secuencias utilizaremos la herramienta **SeqSero2**.

Primero, se debe activar el ambiente específico de la herramienta:
```
conda deactivate
source activate seqsero2_env
```

Cree la carpeta **3-seqsero2**, donde se almacenarán los resultados de salida:
```
mkdir 3-seqsero2
```

El siguiente comando ejecutará **SeqSero2** para cada par de archivos de secuencias filtradas, generando los resultados correspondientes en la carpeta **3-seqsero2**:

```
for f in ../3-Practica_Ensamblaje/1-trim_galore/*_1.fastq.gz ;
do
base=$(basename ${f%_1.fastq.gz})
SeqSero2_package.py -t 2 -m k -i $f ${f%_1.fastq.gz}_2.fastq.gz -p 4 -d 3-seqsero2/${base};
done
```

Una vez finalizada la ejecución, en la carpeta **3-seqsero2** encontrará una subcarpeta para cada muestra. Dentro de cada subcarpeta se encuentra el archivo **SeqSero_result.txt**.

Ábralo y revíselo; contendrá información similar a la mostrada en la siguiente imagen:

![SeqSero2](https://raw.githubusercontent.com/mecalderon/Tutorial_Panama/main/Figures/SeqSero2.png)


### **Ejecución de AMRFinderPlus**

Para identificar genes de resistencia a antibióticos a partir de los genomas ensamblados, utilizaremos la herramienta **AMRFinderPlus**.

Active el ambiente donde se encuentra instalada la herramienta:
```
conda deactivate
source activate microenv
```

Cree la carpeta **4-amrfinderplus** para guardar los archivos de salida:
```
mkdir 4-amrfinderplus
```

Ejecute **AMRFinderPlus** con el siguiente comando:
```
for f in ../3-Practica_Ensamblaje/3-unicycler/* ;
do
base=$(basename $f)
amrfinder -O Salmonella -n $f/${base}.fasta -o 4-amrfinderplus/${base}_amrfinderplus.tsv;
done
```

Este comando ejecuta **AMRFinderPlus** para cada ensamblaje generado previamente y guarda los resultados en la carpeta **4-amrfinderplus**, generando un archivo de salida para cada muestra (**XXXX_amrfinderplus.tsv**) con los genes de resistencia identificados.
