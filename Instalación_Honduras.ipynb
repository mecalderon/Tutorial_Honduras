#**Documentación de instalación de herramientas para el taller**

Este cuaderno documenta los procedimientos necesarios para la instalación de las diferentes herramientas y la descarga de las bases de datos requeridas para el desarrollo del taller.

#**Instalación de CONDA**

**Descripción:**
Conda es una herramienta de código abierto para la gestión de paquetes y entornos. Permite instalar múltiples versiones de paquetes de software junto con sus dependencias, y facilita el cambio entre diferentes entornos de manera sencilla.

La instalación de Conda se debe realizar siguiendo las instrucciones oficiales para sistemas operativos Linux mediante línea de comandos ([CONDA-Linux](https://www.anaconda.com/docs/getting-started/miniconda/install#linux-2)).

**Instalación de CONDA:**

1. Crear una carpeta miniconda:

```
mkdir -p ~/miniconda3
```

2. Descargar el script de instalación de Miniconda:

```
wget https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh -O ~/miniconda3/miniconda.sh
```

3. Correr el script de instalación:

```
bash ~/miniconda3/miniconda.sh -b -u -p ~/miniconda3
```

4. Eliminar el script de instalación:

```
rm ~/miniconda3/miniconda.sh
```


#**Instalación de herramientas para el taller**

Para instalar las herramientas necesarias, se debe ejecutar el siguiente script en Bash:
```
bash installing.sh
```
Este script creará los ambientes **gubbienvs**, **microenv**, **ariba_env** y **srst2_env**, los cuales contienen las dependencias requeridas para el taller.

#**Descarga de bases de datos para el taller**

A continuación, se describen los comandos necesarios para descargar y preparar las bases de datos requeridas durante el taller.

1. **Descarga de la base de datos de Kraken2**

Crear una carpeta para almacenar la base de datos:
```
mkdir Kraken_DB

cd Kraken_DB
```

La base de datos estándar de Kraken2 puede descargarse con el siguiente comando:
```
wget https://genome-idx.s3.amazonaws.com/kraken/k2_standard_08_GB_20260226.tar.gz
```

Descomprimir la base de datos:
```
tar -xvzf k2_standard_08_GB_20260226.tar.gz
```

2. **Descarga y preparación de la base de datos CARD para ARIBA**

Activar el ambiente de ariba:
```
conda activate ariba_env
```

Crear un directorio para la base de datos de ARIBA:
```
mkdir ARIBA_DB
cd ARIBA_DB
```

Ejecute los siguientes comandos para descargar y preparar la base de datos CARD:
```
ariba getref card card.out
ariba prepareref -f card.out.fa -m card.out.tsv card.out
```

3. **Descarga y preparación de la base de datos ResFinder para ARIBA**

Ejecute los siguientes comandos para descargar y preparar la base de datos ResFinder:
```
ariba getref resfinder resfinder.out
ariba prepareref -f resfinder.out.fa -m resfinder.out.tsv resfinder.out
```

4. **Descarga y preparación de la base de datos VirulenceFinder para ARIBA**

Ejecute los siguientes comandos para descargar y preparar la base de datos VirulenceFinder:
```
ariba getref virulencefinder vfd.out
ariba prepareref -f vfd.out.fa -m vfd.out.tsv vfd.out
```

5. **Descarga de base de datos de MLST**

Ejecute el siguiente comando para descargar la base de MLST para Salmonella enterica:
```
ariba pubmlstget 'Salmonella enterica' get_mlst
```

Renombrar base de datos:
```
mv get_mlst/ref_db/ get_mlst/Salmonella_enterica_MLSTdb
```
